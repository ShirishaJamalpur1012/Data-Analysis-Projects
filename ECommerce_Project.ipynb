```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_dark"
```

```python
data = pd.read_csv("Sample - Superstore.csv", encoding='latin-1') #using Latin so that it can read all the data inclusing special chars
```

```python
data.head() #head() provides the top most cells in the data, by deafult the values is 5 if we need more cells we have to explicitly provide the count
```

```python
data.describe() #This function provides the basic description of integer columns
```

```python
data.info() #This function gives the complete information of the columns along with their data types, null values
```

**Type Conversion**

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

**Creating New Columns**

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of Week'] = data['Order Date'].dt.dayofweek
```

**Getting in between data**

```python
data[10:20]
```

**Monthly Sales Analysis**

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
fig = px.line(sales_by_month, x='Order Month', y='Sales', title='Monthly Sales Analysis')
fig.show()
```

**Sales by Category**

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
fig = px.pie(sales_by_category,
             names = 'Category',
             values = 'Sales',
             hole = 0.5,
             color_discrete_sequence=px.colors.qualitative.Prism)

fig.update_traces(textposition='inside',textinfo = 'percent+label')
fig.update_layout(title_text='Sales By Category',title_font=dict(size=24),title_font_family='Bold')

fig.show()
```

**Sales by SubCategory**

```python
Sales_By_SubCategory = data.groupby(['Category','Sub-Category'])['Sales'].sum().reset_index()
fig = px.bar(Sales_By_SubCategory,
             x='Sub-Category',
             y = 'Sales',
             title='Sales by SubCategory',
             color = 'Category',
             barmode = 'group')
fig.show()
```

**Analyse Monthly Profit from Sales**

```python
Monthly_Profit_From_Sales = data.groupby('Order Month').agg({'Sales':'sum','Profit':'sum'}).reset_index()
fig = px.bar(Monthly_Profit_From_Sales,
              x = 'Order Month',
              y = ['Sales','Profit'],
              barmode = 'group')
fig.update_layout(xaxis=dict(tickmode='linear'))
fig.show()
```

**Analyse Profit by Category and SubCategory**

```python
Profit_by_Category_SubCategory = data.groupby(['Category','Sub-Category'])['Profit'].sum().reset_index()
fig = px.bar(Profit_by_Category_SubCategory,
             x = 'Category',
             y = 'Profit',
             color = 'Sub-Category',
             barmode = 'group')
fig.show()
```

**Analyse Sales and Profit by Customer Segment**

```python
profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
fig = px.bar(profit_by_segment,
             x = 'Segment',
             y = ['Sales','Profit'])
fig.update_layout(barmode='group')
fig.show()
```

**Analyse Sales to Profit Ratio**

```python
Sales_to_Profit = data.groupby('Category').agg({'Sales':'sum','Profit':'sum'}).reset_index()
Sales_to_Profit['Sales_to_Profit_Ratio'] = Sales_to_Profit['Sales']/Sales_to_Profit['Profit']
fig = px.pie(Sales_to_Profit,
             names = 'Category',
             values = 'Sales_to_Profit_Ratio',
             hole = 0.5)
fig.update_traces(textposition='inside',textinfo = 'percent+label')
fig.show()
```
